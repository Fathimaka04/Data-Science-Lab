```python
# SVD

from scipy.linalg import svd
matrix=np.arange(1,19).reshape(6,3)
```

```python
U,s,VT=svd(matrix)
U
```

```python
s
```

```python
VT
```

```python
U,s,VT=svd(matrix,full_matrices=True)
U
```

```python
U,s,VT=svd(matrix,full_matrices=False)
U
```

```python
from numpy import diag
from numpy import dot
matrix=(U@np.diag(s)@VT)
matrix
```
