```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
plt.plot(x,y,marker=".", markersize=10)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X-Y graph")
plt.show()
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'b--')
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g^')
```

```python
#  using object oriented 
import matplotlib.pyplot as plt
fig=plt.figure()
axes=fig.add_axes([0.5,0.9,0.99,0.89])
axes.plot(x,y,'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('X-Y Graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.7,0.6,0.9,0.9])
axes2=fig.add_axes([0.8,0.99,0.4,0.4])
# Main Figure axes1
axes1.plot(x,y,'b')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('X-Y Graph')
# insert Figure axes1
axes2.plot(y,x,'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('X-Y Graph')
```

```python
fid,axes=plt.subplots()
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('X_Y Graph')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.text(0.5,1.0,'X-Y Graph')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)

plt.plot(
    x, x + 3,
    color='b',      
    lw=3,
    linestyle='dashed',
    marker='.',
    markersize=12,
    markeredgecolor='green',
    markeredgewidth=5,
    markerfacecolor='red'
)

plt.plot(
    x, x + 4,
    color='g',         
    lw=4,
    linestyle='dashed',
    marker='o',
    markersize=10,
    markeredgecolor='blue',
    markeredgewidth=5,
    markerfacecolor='green'
)

plt.show()
```

```python
import seaborn as sns
print(sns.__version__)
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
tips.head()
```

```python
sns.displot(tips["tip"])   #displot-Displays the distribution of a numerical variable.
```

```python
sns.distplot(tips['total_bill'],kde=True,bins=30)



"""
distplot() is the old (deprecated) version of Seaborn's distribution plot.

It was used to display a histogram and optionally 
a KDE (Kernel Density Estimate) curve on the same plot."""
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')

"""  jointplot() in Seaborn is used to show the relationship between two numerical variables. It combines:

    A main plot (scatter, histogram, KDE, etc.) showing the relationship.
    A distribution plot for the x variable (top).
    A distribution plot for the y variable (right).
"""
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.rugplot(tips['total_bill'])

"""
rugplot() is a Seaborn function that draws
a small tick mark (rug) for every data point along an axis.
eG:marks = [10, 20, 20, 30, 40]
|    ||    |     |
10   20    30    40
Each | represents one data point.
Two | at 20 mean there are two observations with the value 20.
"""
```
