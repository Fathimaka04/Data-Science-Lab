```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
plt.plot(x,y,marker=".", markersize=10)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X-Y graph")
plt.show()
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'b--')
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g^')
```

```python
#  using object oriented 

fig=plt.figure()
axes=fig.add_axes([0.5,0.9,0.99,0.89])
axes.plot(x,y,'b')
```
