```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
plt.plot(x,y,marker=".", markersize=10)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X-Y graph")
plt.show()
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'b--')
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g^')
```

```python
#  using object oriented 
import matplotlib.pyplot as plt
fig=plt.figure()
axes=fig.add_axes([0.5,0.9,0.99,0.89])
axes.plot(x,y,'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('X-Y Graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.7,0.6,0.9,0.9])
axes2=fig.add_axes([0.8,0.99,0.4,0.4])
# Main Figure axes1
axes1.plot(x,y,'b')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('X-Y Graph')
# insert Figure axes1
axes2.plot(y,x,'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('X-Y Graph')
```

```python
fid,axes=plt.subplots()
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('X_Y Graph')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.text(0.5,1.0,'X-Y Graph')
```
